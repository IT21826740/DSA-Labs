```python
import pandas as pd
from apyori import apriori
```

```python
data = pd.read_csv("movie_dataset.csv", header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
records = []
for i in range(1, 7501):
   records.append([str(data.values[i, j]) for j in range(0, 20)])
```

```python
association_rules = apriori(records, min_support=0.01)
```

```python
association_results = list(association_rules)
```

```python
association_results
```

```python
association_rules = apriori(records, min_support=0.01, min_confidence=0.5)
```

```python
association_results = list(association_rules)
```

```python
association_results
```
